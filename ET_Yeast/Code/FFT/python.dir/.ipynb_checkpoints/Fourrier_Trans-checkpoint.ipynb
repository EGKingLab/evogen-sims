```python
import pandas as pd
import cv2, math,json,geopy.distance, os, random
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import fftpack
from sklearn import metrics
```

```python
class Freq:
    def __init__ (self, path):
        self.path = path

    def read_my_files(self):
        mydata = pd.read_csv(self.path)
        return mydata
    
    def myseq(self):
        df = self.read_my_files()
        return df
    
    def visual(self, mydata):
        #return mydata.plot(x="Generation", y="Frequency",  kind="line")

        fig, ax = plt.subplots(figsize=(10, 6))
        sns.lineplot(data=mydata, x='Generation', y='Frequency', hue='Position', ax=ax, legend=False)
        # Remove the left and bottom spines
        ax.spines['left'].set_visible(False)
        ax.spines['bottom'].set_visible(False)

        # Remove the ticks
        ax.tick_params(left=False, bottom=False)

        # Remove the axis labels
        plt.xlabel('')
        plt.ylabel('')

        # Remove the top and right spines using sns.despine()
        sns.despine(top=True, right=True)

        # Remove the x and y tick labels
        ax.set_xticklabels([])
        ax.set_yticklabels([])
        
        plt.show()
```

```python
geno_file_f = pd.read_csv("/home/etb68/YeastProj.dir/evogen-sims/ET_Yeast/output.dir/Selection_Models/CSFP.1.dir/genomeCSFP_A1_100_0.5.csv")
splt_data = geno_file_f.groupby('Position')

# Define a function to apply the Fourier transformation on a group
def fft_group(group):
  # Get the frequency values as a numpy array
  freq = group["Frequency"].values
  # Compute the Fourier transform of the frequency values
  fft_freq = fftpack.fft(freq)
  # Add the Fourier transform as a new column to the group
  group["power"] = np.abs(fft_freq) #Signal power
  group["fft_freq"] = fftpack.fftfreq((np.abs(fft_freq)).size, d = 1)#Signal frequency. Signal = allele freq
  
  #pos_mask = list(np.where(group["fft_freq"] > 0)[0])
  pos_mask = np.where(group["fft_freq"] > 0) #Freq greater than zero
  peak_indx = group["power"][pos_mask].argmax()
  #peak_indx = group.loc[pos_mask, "power"].argmax()
  peak_freq = group["fft_freq"][power[pos_mask].argmax()]

  lf_signal = fftpack.ifft(fft_freq) #Inverse fourier transformation. It takes us back to the signal
  #return lf_signal, sample_freq

  # Return the modified group
  return group, lf_signal

# Apply the function to each group and concatenate the results
rev_data = pd.concat(fft_group(group) for _, group in splt_data)
rev_data.tail(6)
```
